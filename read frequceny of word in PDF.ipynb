```python
import PyPDF2,re,string
frequency={}
pdfFileObj=open('D:\\Sumi\\课本\\43-47课本\\2109 春3 二年级学生用书  L43-47.pdf','rb')
pdfReader=PyPDF2.PdfFileReader(pdfFileObj)
for pageNum in range(5,18):
    pageObj=pdfReader.getPage(pageNum)

text_string=pageObj.extractText().lower()
match_pattern=re.findall(r'\b[a-z]{6,15}\b',text_string)

for word in match_pattern:
    count=frequency.get(word,0)
    frequency[word]=count+1

l=[]
for a,b in frequency.items():
    l.append((b,a))
l.sort(reverse=True)

for words,frequency[words] in l:
    print(words,frequency[words])
```
